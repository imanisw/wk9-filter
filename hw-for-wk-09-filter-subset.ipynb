```python
## CHALLENGE (as in you have to google this)
## How many people were killed in Queens in accidents?
fil_boro = df_nyc_accidents["BOROUGH"] == "QUEENS"
fil_killed = df_nyc_accidents["NUMBER OF PERSONS KILLED"] 

df_challenge = df_nyc_accidents[fil_boro & fil_killed].sum
df_challenge
```

```python
## Same
## how many cyclists were killed in Queens?
fil_boro = df_nyc_accidents["BOROUGH"] == "QUEENS"
fil_cyclist = df_nyc_accidents["NUMBER OF CYCLIST KILLED"] 

df_nyc_accidents[fil_boro & fil_cyclist].sum
```

```python
## Filter and subset 
## create a dataset for Manhattan that involved taxi cabs as the primary vehicle cause

fil_man = df_nyc_accidents["BOROUGH"] == "MANHATTAN"
fil_car = df_nyc_accidents["CONTRIBUTING FACTOR VEHICLE 1"] == "TAXI CAB"

df_man = df_nyc_accidents[fil_man & fil_car]
df_man

```
